```python
def cost(b):
    return (b - 4) ** 2
```

```python
def num_slope(b):
    h = .0001
    return (cost(b+h) - cost(b))/h
```

```python
num_slope(3.00001)
```

```python
def slope(b):
    return 2 * (b - 4)
```

```python
slope(5)
```

```python
#let's apply the slope rule to minimize the cost functio
b = -20
```

```python
b = b - .1 * slope(b)
print(b)
```

```python
#training loop by minimizing cost through minimizing errors
b = 5
for i in range(100):
    b = b - .1 * slope(b)
    print(b)
```
